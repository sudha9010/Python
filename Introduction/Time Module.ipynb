```python
The time module in Python provides various functions for working with time-related operations. 
It allows you to measure time intervals, format dates and times, and perform various other time-related operations. 
```

```python
import time

# Get the current time in seconds since the epoch 
current_time = time.time()
print("Current time:", current_time)

# Pause the program execution for 2 seconds
print("Before sleep")
time.sleep(2)
print("After sleep")

# Get the current local time as a struct_time object
local_time = time.localtime()
print("Local time:", local_time)

# Format the current local time as a string
formatted_time = time.strftime("%Y-%m-%d %H:%M:%S", local_time)
print("Formatted time:", formatted_time)

# Measure the execution time of a piece of code
start_time = time.time()
# Code block to measure the execution time
time.sleep(1)
end_time = time.time()
execution_time = end_time - start_time
print("Execution time:", execution_time, "seconds")
```

```python
import datetime

def get_years_with_august_15_friday():
    current_year = datetime.datetime.now().year
    years = range(1, current_year + 1)

    friday_years = []
    for year in years:
        if datetime.date(year, 8, 15).weekday() == 4: #check for aug 15th - friday
            friday_years.append(year)

    return friday_years


august_15_friday_years = get_years_with_august_15_friday() #years


print("Years on which August 15th is a Friday:")
for year in august_15_friday_years:
    print(year)
```

```python
import calendar

def print_july_calendars():
    current_year = calendar.datetime.datetime.now().year
    years = range(1, current_year + 1)

    for year in years:
        
        print(f"{year} - July")
        print(calendar.month(year, 7))
        print()


print_july_calendars()
```

```python
import calendar

# Get the current year
current_year = calendar.datetime.datetime.now().year

# Define the starting year (first century)
starting_year = 1

# Iterate over each year from the starting year to the current year
for year in range(starting_year, current_year + 1):
    # Check if August 15th is a Friday
    if calendar.weekday(year, 8, 15) == calendar.FRIDAY:
        # Print the year and month
        print(f"August 15th, {year} is a Friday:")
        # Print the calendar for August of that year
        print(calendar.month(year, 8))
        print()
```

```python
import calendar
from colorama import Fore, Back, Style

def print_july_calendar():
    current_year = calendar.datetime.datetime.now().year
    start_year = 1997

    for year in range(start_year, current_year + 1):
        # Print the year
        print(f"{year}")
        
        # Get the calendar for the year
        cal = calendar.monthcalendar(year, 7)

        # Print the calendar month of July
        print(calendar.month_name[7])
        print("Mo Tu We Th Fr Sa Su")

        for week in cal:
            for day in week:
                # Check if the day belongs to July
                if day != 0:
                    # Check if it's July 18th and colorize it
                    if day == 18:
                        print(Back.RED + Fore.WHITE + f"{day:2}", end=" ")
                    else:
                        print(f"{day:2}", end=" ")
                else:
                    print("  ", end=" ")
            print()
        
        print(Style.RESET_ALL)
        print()

# Call the function to print July calendars
print_july_calendar()
```
