```python
import numpy as np
import matplotlib.pyplot as plt

data = [[10, 20, 30, 40, 50],
        [65, 15, 25, 35, 45],
        [23, 21, 17, 19, 22]]

X = np.arange(5)

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X + 0.00, data[0], color='b', width=0.25)
ax.bar(X + 0.25, data[1], color='r', width=0.25)
ax.bar(X + 0.50, data[2], color='g', width=0.25)

plt.show()
```

The code begins by importing the necessary libraries: NumPy and Matplotlib. NumPy is a popular numerical computing library, and Matplotlib is a plotting library that allows us to create various types of visualizations.

The variable data is defined as a 2D list. It represents the data points for the bar chart. Each sublist within data corresponds to a different set of bars on the chart.

The variable X is created using np.arange(5). It is an array representing the x-coordinates for each bar in the chart. In this case, there are five bars, so X will be an array with values [0, 1, 2, 3, 4].

A new figure is created using fig = plt.figure(). This figure will contain the bar chart.

An axes object is added to the figure using ax = fig.add_axes([0, 0, 1, 1]). The [0, 0, 1, 1] argument specifies the position and size of the axes within the figure. The values range from 0 to 1, where (0, 0) represents the bottom-left corner and (1, 1) represents the top-right corner of the figure.

The ax.bar() function is called three times to create the bars on the chart. Each call specifies the x-coordinates (X + 0.00, X + 0.25, X + 0.50), the corresponding data points (data[0], data[1], data[2]), the color ('b', 'r', 'g'), and the width of the bars (0.25).

Finally, plt.show() is called to display the chart.

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[10, 20, 30, 40, 50],
        [65, 15, 25, 35, 45],
        [23, 21, 17, 19, 22]]

# Calculate the total for each group
totals = np.sum(data, axis=1)

# Create labels for each group
labels = ['Group 1', 'Group 2', 'Group 3']

# Plot the pie chart
plt.pie(totals, labels=labels, autopct='%1.1f%%')

# Add a title
plt.title('Pie Chart')

# Display the chart
plt.show()
```

In this code, we first calculate the total for each group by summing the values along the rows using np.sum(data, axis=1). Then, we create labels for each group. We pass the totals and labels arrays to plt.pie() function to generate the pie chart. The autopct='%1.1f%%' argument is used to display the percentage values on the chart. Finally, we add a title to the chart using plt.title() and display it using plt.show().
